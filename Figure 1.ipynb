```python
import scipy as sp, numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.pyplot import figure
import pandas as pd
```

```python
conda_forge = 2015
bioconda = 2015
# might need to change linux/home brew
linuxhomebrew = 2009
debian_med = 1993
singularity = 2016
spack = 2013
easy_build = 2012
pip = 2008
biocontainers = 2015
gentoo_science = 2000
biolinux = 2014
gnu_guix = 2013
biobuilds = 2016
kubernetes = 2014
yum = 1999
apt_get = 1998
vagrant = 2010
```

```python

labels = ["Debian Med", "apt-get", "yum", "Gentoo Science", "pip", "Linux/Home Brew", "Vagrant", "EasyBuild", "GNU Guix", "Spack", "Biolinux", "conda forge", "Bioconda", "Biobuilds"]
manager_list = [conda_forge, bioconda, linuxhomebrew, debian_med,spack, easy_build, pip, gentoo_science, biolinux, gnu_guix, biobuilds, yum, apt_get, vagrant]
manager_list2 = np.sort(manager_list)

yaxis = np.arange(1,len(manager_list)+1)

colors = ['deepskyblue','deepskyblue','deepskyblue','deepskyblue','deepskyblue','deepskyblue','deepskyblue','deepskyblue', 'deepskyblue','deepskyblue','deepskyblue','deepskyblue','deepskyblue','deepskyblue']

plt.scatter(manager_list2, yaxis, s = 100, c = colors, marker =(8,1,0),alpha = .5);
#color coded the figure package manager vs containers

for i in range(len(manager_list2)):
    plt.annotate(labels[i],(manager_list2[i], yaxis[i]))
    plt.axhline(y=i+1, alpha = .4, linewidth = .3, color = 'slateblue')
    
plt.box(on=None)  
plt.show()
#plt.axis('off')
```
